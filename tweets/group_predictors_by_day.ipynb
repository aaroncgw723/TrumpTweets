```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('white')
```

```python
predictors = pd.read_csv('../tweets/cleaned_tweets_text')
columns = predictors.columns.values

predictors['clean_text'] = predictors['clean_text'].astype(str)

pred_by_day = predictors.groupby([predictors.year.values,predictors.month.values, predictors.day.values]
).agg(
        {
            'clean_text': sum,
            'year':'first',
            'month':'first',
            'day':'first',
        }
    )

pred_by_day['Date'] = [f"{int(a[0]):d}-{int(a[1]):02d}-{int(a[2]):02d}" for a in zip(pred_by_day['year'],pred_by_day["month"],pred_by_day["day"])]
pred_by_day.head(10)
```

```python
pred_by_day.to_csv('../tweets/cleaned_tweet_text_by_day')
```

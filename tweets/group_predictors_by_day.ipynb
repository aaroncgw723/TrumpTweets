```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('white')
```

```python
predictors = pd.read_csv('../tweets/cleaned_tweets')
columns = predictors.columns.values

pred_by_day = predictors.groupby([predictors.year.values,predictors.month.values, predictors.day.values]
).agg(
        {
            'num_words': sum,
            'num_hashtags': sum,
            'is_retweet': sum,
            'favorite_count': sum,
            'retweet_count': sum,
            'num_mentions': sum,
            'keyword_China': sum,
            'keyword_Russia': sum,
            'keyword_Mexico': sum,
            'keyword_Canada': sum,
            'keyword_Korea': sum,
            'keyword_Clinton': sum,
            'keyword_Democrats': sum,
            'keyword_Republicans': sum,
            'keyword_himself': sum,
            'keyword_family': sum,
            'keyword_government': sum,
            'keyword_media': sum,
            'keyword_states': sum,
            'keyword_social_media': sum,
            'keyword_Obama': sum,
            'keyword_FBI': sum,
            'keyword_rivals': sum,
            'keyword_MiddleEast': sum,
            'keyword_Ukraine_scandal': sum,
            'keyword_economy': sum,
            'keyword_disaster': sum,
            'year':'first',
            'month':'first',
            'day':'first',
        }
    )

pred_by_day['Date'] = [f"{int(a[0]):d}-{int(a[1]):02d}-{int(a[2]):02d}" for a in zip(pred_by_day['year'],pred_by_day["month"],pred_by_day["day"])]
pred_by_day.head(10)
```

```python
pred_by_day.to_csv('../tweets/cleaned_tweets_by_day')
```

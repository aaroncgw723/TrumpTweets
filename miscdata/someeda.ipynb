```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
gold = pd.read_csv("golddata.csv")
oil = pd.read_csv("oildata.csv")
btc = pd.read_csv("bitcoindata.csv")
forex_cad = pd.read_csv("usd_cad_exchangerate.csv")
forex_cny = pd.read_csv("usd_cny_exchangerate.csv")
forex_krw = pd.read_csv("usd_krw_exchangerate.csv")
forex_krw = forex_krw.iloc[forex_krw.index[forex_krw.iloc[:,1]!='.'].tolist(), :]
forex_mxn = pd.read_csv("usd_mxn_exchangerate.csv")
forex_rub = pd.read_csv("usd_rub_exchangerate.csv")
```

```python
scaler = MinMaxScaler()

goldprices = gold[["price"]].values.astype(float)
goldnew = scaler.fit_transform(goldprices)
goldnew = pd.DataFrame(goldnew)

oilprices = oil[["Price"]].values.astype(float)
oilnew = pd.DataFrame(scaler.fit_transform(oilprices))

#normalizing the variables
```

```python
plt.plot(goldnew, label="gold")
plt.plot(oilnew, label="oil")
plt.xlabel("# of days starting from Jan 1 2017")
plt.ylabel("Normalized price")
plt.legend(loc="upper left")


```

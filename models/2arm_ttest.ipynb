```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('white')

from scipy.stats import ttest_ind
```

```python
df = pd.read_csv('rub_ttest_data.csv', sep=";")
df.columns = ["model_name", "t", "p", "l", "r_5", "r", "r_95"]
df.describe()
```

```python
fig = plt.figure(figsize=(10,7))
sns.distplot(df['l'], ax=fig.gca(), label='Model Loss')
sns.distplot(df['r'], ax=fig.gca(), label='Y-Randomized Loss')

plt.title('Model Performance Against Y-Randomized Control')
plt.xlabel('Loss')
plt.ylabel('Abundance')

plt.legend(loc='best')
plt.show()
```

```python
ttest_ind(df['l'], df['r'])
```
